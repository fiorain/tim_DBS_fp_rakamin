```python
def date_correction(x,verbose=True):
    date_split = x.split("/")
    count_correction = 0
    if len(date_split[-1]) == 2:
        corrected_date = "19"+date_split[-1]
        final_date = "/".join(date_split[:-1]+[corrected_date])
        count_correction+=1
    else:
        final_date = x
    return(final_date)
```

```python
df['new_dob'] = df['DOB'].map(lambda x: datetime.strptime(date_correction(x),"%m/%d/%Y"))
```

```python
df['tes_date_of_hire'] = df['DateofHire'].map(lambda x: datetime.strptime(date_correction(x), "%m/%d/%Y"))
```
